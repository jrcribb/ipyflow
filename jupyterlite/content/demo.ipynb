# IPyflow — reactive Python notebooks, in your browser

A live [JupyterLite](https://jupyterlite.readthedocs.io) demo of [ipyflow](https://github.com/ipyflow/ipyflow) — no install, everything runs in your browser via Pyodide.

**Try it:** run the cells top-to-bottom (Run ▸ Run All Cells). The first run takes ~30–60s while Python loads in the browser.

The first cell activates ipyflow with `%load_ext ipyflow` (it comes pre-installed). After that:
1. **Re-run the `x = 1` cell** — ipyflow highlights `y = x + 1` (and its dependents) as **stale / ready to re-run**.
2. Press **Ctrl/Cmd+Shift+Enter** on a cell to reactively re-execute it *and everything downstream* that depends on it.

```python
%load_ext ipyflow
```

```python
x = 1
```

```python
y = x + 1
```

```python
z = y * 10
```

```python
print(f'x={x}, y={y}, z={z}')
```
